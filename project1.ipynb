```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)  
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
From above graphs we can say that most of the buyers are female 
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)  
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
From the above graphs we cans see that most of the buyers are from the age group of 26-35 and the are mostly female
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='Orders',y='State')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='Amount',y='State')
```

```python
From above we can say that most of the orders are placed from Uttar Pradesh , Maharashtra and Karnataka
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(8,5)})
for bars in ax.containers:
  ax.bar_label(bars)  
```

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
  ax.bar_label(bars)  
```

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(8,5)})
for bars in ax.containers:
  ax.bar_label(bars)  
```

```python
sales_prod=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_prod,x='Amount',y='Product_Category')
```
